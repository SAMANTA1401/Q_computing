What is Quantum Teleportation?
Quantum teleportation doesn’t physically move a qubit but transfers its quantum state (e.g., 
α∣0⟩+β∣1⟩) from one location (Alice) to another (Bob) using:

An entangled pair of qubits (a Bell state)

Classical communication (2 bits).

Local quantum operations.

It leverages entanglement to "teleport" the state, not the particle itself, and was proposed by Bennett et al. in 1993.
