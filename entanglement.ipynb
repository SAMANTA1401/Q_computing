What Does "Collapse" Mean?

In quantum mechanics, a quantum system (e.g., a qubit or entangled qubits) exists in a superposition of all possible states until it’s measured. Measurement "collapses" this superposition into a single definite state, chosen probabilistically based on the amplitudes of the state’s components. For entangled systems, this collapse affects all entangled particles, not just the one measured.

Entanglement and Collapse
When qubits are entangled, their states are no longer independent—they share a single joint quantum state. Measuring one qubit doesn’t just determine its own state; it instantly fixes the states of all entangled qubits, no matter how far apart they are. This is what makes entanglement "spooky" and powerful.

Quantum Twist: Before measurement, the qubits don’t "know" their states; the superposition is real. Measuring one forces a decision that ripples to the other.
