What is Superdense Coding?

Superdense coding (proposed by Bennett and Wiesner in 1992) allows Alice to send two classical bits (00, 01, 10, 11) to Bob by physically transmitting only one qubit, provided they pre-share an entangled pair. This doubles the classical capacity of a qubit (normally one bit) using entanglement and a shared quantum resource.
